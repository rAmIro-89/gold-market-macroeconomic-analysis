### Limpieza y conversión de la columna 'date'

### Conversión de `date` a `datetime`
La columna `date` viene en formatos como “Q1 14”.  
Se normaliza a `datetime64[ns]` usando el **fin de cada trimestre**, lo que facilita el análisis cronológico y la interpolación temporal.


```python
import re
from datetime import datetime

# limpiar nombres de columnas
df.columns = df.columns.str.strip()

def parse_quarter_or_year(x):
    if pd.isna(x): return pd.NaT
    if isinstance(x, (pd.Timestamp, datetime)): return pd.Timestamp(x)
    s = str(x).strip()
    m = re.match(r"^Q\s*([1-4])\s*['\-]?\s*(\d{2,4})$", s, flags=re.I)
    if m:
        q = int(m.group(1)); y = int(m.group(2))
        if y < 100: y += 2000
        return pd.Period(f"{y}Q{q}", freq="Q").to_timestamp(how="end")
    m = re.match(r"^(\d{4})$", s)
    if m:
        y = int(m.group(1))
        return pd.Timestamp(year=y, month=12, day=31)
    return pd.to_datetime(s, errors="coerce")

df['date_parsed'] = df['date'].apply(parse_quarter_or_year)
bad = df[df['date_parsed'].isna()]['date'].unique()
print("Chequeo fechas no parseadas (muestra):", bad[:10])

df['date'] = df['date_parsed']
df.drop(columns=['date_parsed'], inplace=True)
df = df.sort_values('date').reset_index(drop=True)
df['date'].dtype
```

### Tratamiento de valores faltantes
Las series son trimestrales; para preservar la continuidad sin eliminar períodos, se aplica **interpolación temporal**:
1) Se fuerza conversión a numérico (`errors='coerce'`) para asegurar el cálculo.  
2) Se interpola con `method='time'` sobre las filas con fecha válida.  
3) Se reinyectan los valores interpolados a la tabla original.  


```python
# Diagnóstico antes
print("Nulos antes:\n", df.isnull().sum().sort_values(ascending=False), "\n")
```

```python
# Interpolación temporal sobre filas con fecha válida
df_with_date = df[df['date'].notna()].copy().set_index('date')

# Forzar numéricos para permitir interpolación
for c in df_with_date.columns:
    if c != df_with_date.index.name:
        df_with_date[c] = pd.to_numeric(df_with_date[c], errors='coerce')

num_cols = df_with_date.select_dtypes(include='number').columns.tolist()

if num_cols:
    df_with_date[num_cols] = df_with_date[num_cols].interpolate(
        method='time', limit_direction='both'
    )

# Reinyectar solo columnas numéricas
df_interp = df_with_date.reset_index()[['date'] + num_cols]
# preservo orden original
cols_order = df.columns.tolist()
df = df.merge(df_interp, on='date', how='left', suffixes=('', '_interp'))
for c in num_cols:
    df[c] = df[f'{c}_interp'].combine_first(df[c])
    df.drop(columns=[f'{c}_interp'], inplace=True)
df = df[cols_order]

# Verificación después
print("Nulos después:\n", df.isnull().sum().sort_values(ascending=False))
```

## 3) EDA – Análisis univariado
Se exploran distribuciones y evolución temporal de variables clave.

Se seleccionaron cinco variables clave para el análisis univariado.  
El objetivo es describir su comportamiento de manera independiente, antes de relacionarlas con el oro.

```python
import matplotlib.pyplot as plt

# Serie temporal del oro
plt.figure()
plt.plot(df['date'], df['gold_price (USD/oz)'])
plt.title('Evolución del precio del oro (USD/oz)')
plt.xlabel('Fecha'); plt.ylabel('USD/oz'); plt.tight_layout()

# Histograma de inflación (YoY)
plt.figure()
df['us_inflation (%, YoY CPI)'].hist(bins=10)
plt.title('Distribución de inflación (YoY, %) - EE.UU.')
plt.xlabel('% interanual'); plt.ylabel('Frecuencia'); plt.tight_layout()

# Boxplot del petróleo
plt.figure()
df['oil_price (USD/bbl)'].plot(kind='box')
plt.title('Boxplot del precio del petróleo (USD/bbl)')
plt.ylabel('USD/barril'); plt.tight_layout()

# Serie temporal del VIX (índice de volatilidad)
plt.figure()
plt.plot(df['date'], df['vix_index'])
plt.title('Evolución del VIX (índice de volatilidad)')
plt.xlabel('Fecha'); plt.ylabel('VIX'); plt.tight_layout()

# Serie temporal del S&P500
plt.figure()
plt.plot(df['date'], df['sp500_index (points)'])
plt.title('Evolución del índice S&P500 (puntos)')
plt.xlabel('Fecha'); plt.ylabel('Puntos'); plt.tight_layout()
```

1. **Precio del oro (`gold_price (USD/oz)`) – Serie temporal (línea)**  
   - *Por qué esta variable:* es el eje central del proyecto y muestra la tendencia del oro entre 2014–2024.  
   - *Por qué este gráfico:* una línea temporal permite observar la evolución continua, tendencias y posibles rupturas.

2. **Inflación de EE.UU. (`us_inflation (%, YoY CPI)`) – Histograma**  
   - *Por qué esta variable:* la inflación es un determinante macroeconómico clave para entender la demanda de activos refugio como el oro.  
   - *Por qué este gráfico:* el histograma permite analizar la **distribución de frecuencias**, mostrando valores típicos y dispersión.

3. **Precio del petróleo (`oil_price (USD/bbl)`) – Boxplot**  
   - *Por qué esta variable:* el petróleo es un commodity estratégico que puede influir en expectativas de inflación y en la relación oro-dólar.  
   - *Por qué este gráfico:* el boxplot facilita la detección de **valores atípicos** y la medición de la variabilidad en una serie muy volátil.

4. **Índice de volatilidad VIX (`vix_index`) – Serie temporal (línea)**  
   - *Por qué esta variable:* mide la incertidumbre de los mercados financieros; el oro suele apreciarse en contextos de alta volatilidad.  
   - *Por qué este gráfico:* la línea temporal permite identificar picos de volatilidad y compararlos más adelante con movimientos del oro.

5. **Índice bursátil S&P500 (`sp500_index (points)`) – Serie temporal (línea)**  
   - *Por qué esta variable:* representa el desempeño del mercado accionario de EE.UU.; el oro suele actuar como cobertura en períodos de caídas bursátiles.  
   - *Por qué este gráfico:* una serie temporal permite observar la dinámica del índice y posibles episodios de crisis o recuperación.



## EDA – Análisis bivariado
Se analiza la relación del oro con: reservas de bancos centrales, índice del dólar y tasas reales.

#### Oro vs reservas de bancos centrales (tonnes) → para ver si acumular reservas se asocia a mayor precio.

#### Oro vs usd_index (2010=100) → para ver la relación inversa clásica entre oro y dólar.
